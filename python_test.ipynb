```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
x = np.linspace(-4,4,9 )
y = np.linspace(-5,5,11)
x_1, y_1 = np.meshgrid(x,y)
z = (3*x_1) + (2*y_1) + 5

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(x_1, y_1, z, color = 'C3')
plt.show(block = False)
plt.savefig('3d_plot_1.jpg')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi

r = 1.0
th = np.linspace(0,(2*pi), 200)
ph = np.linspace(0,(2*pi),200)
u,v = np.meshgrid(th, ph)

x = r * np.cos(u) * np.sin(v)
y = r * np.sin(u) * np.sin(v)
z = r * np.cos(v)

fig = plt.figure()
ax = plt.axes(projection = '3d')
ax.plot_surface(x,y,z, color = 'C2')
plt.show(block = False)
plt.savefig('3d_plot.jpg')
```
